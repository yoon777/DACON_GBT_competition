```python
from bertopic import BERTopic
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/DACON_해커톤/dataset/train.csv')
df
```

```python
# 두 문자열을 합치기
df['text'] = df['제목'] + " " + df['분류']  # 사이에 공백 추가
df['text']
```

```python
list_of_text = df['text'].tolist()
list_of_text
```

```python
from bertopic import BERTopic
from tqdm import tqdm
import torch

# GPU 사용 확인
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# BERTopic 모델 생성
model = BERTopic()

# list_of_text에 대해 fit_transform 진행
with tqdm(total=len(list_of_text), desc="Fitting BERTopic") as pbar:
    topics, probabilities = model.fit_transform(list_of_text)
    pbar.update(len(list_of_text))  # 전체 데이터 수만큼 진행률 업데이트

# 결과 출력
print('각 문서의 토픽 번호 리스트:', len(topics))
print('첫번째 문서의 토픽 번호:', topics[0])

# 주제 정보 출력
topic_info = model.get_topic_info()
print(topic_info)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.preprocessing import OneHotEncoder

# 2. 문서 데이터를 학습/검증 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(df['text'].to_list(), df['분류'].to_list(), test_size=0.2, random_state=42)
test = pd.read_csv('/content/drive/MyDrive/DACON_해커톤/dataset/test.csv')

# 3. BERTopic을 사용하여 주제 모델링
model = BERTopic()
topics_train, _ = model.fit_transform(X_train)

# 4. 학습 데이터의 토픽 번호를 레이블로 사용
X_train_topics = topics_train

# 5. 로지스틱 회귀 모델 학습
classifier = LogisticRegression()
classifier.fit(X_train_topics, y_train)

# 6. 테스트 데이터의 토픽 예측
topics_test, _ = model.transform(X_test)
predictions = classifier.predict(topics_test)

# 7. 분류 성능 평가
print(classification_report(y_test, predictions))
```
